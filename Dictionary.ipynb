## Input value from user and store in list

Take input from user. store in dictionary and then append dictionary to a list

````python
data = []
userAnswer = "y"
while userAnswer != "n":
    studentName = input("Enter your name: ")
    studentEmail = input("Enter your email: ")
    studentClass = input("Enter your class: ")
    
    studentData = {
        "name" : studentName,
        "email" : studentEmail,
        "class" : studentClass
    }
    
    data.append(studentData)
    
    userAnswer = input("Do you want to continue. y/n:")
    
    
print("Your data:", data)
    
    
    






````
